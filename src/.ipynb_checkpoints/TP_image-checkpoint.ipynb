# 2. Load the data dictionnary, add the unprocessed image if needed



```python
def preprocess_image():
    
    #load labels
    labels = pickle.load(open(datapath+"msdi_img_labels.pickle", 'rb'))
    print(type(labels))

    # Load data dictionnary (raw data)
    with open(datapath + 'data.pickle', 'rb') as f:
      data = pickle.load(f)
    #print(type(data))

    #Add unopened images to the data dict
    file_list = os.listdir(datapath + "img")

    file_list = list(set(file_list) - set(data.keys())) # if not empty, there are still images to add to the data dictionnary
    print(len(file_list))

    #data = {}
    for f in file_list:
      data[f] = np.array(Image.open(datapath + "img/" + f))

    # save raw data dictionnary if save ==True
    save = False
    if save :
      with open(datapath + 'data.pickle', 'wb') as f:
        # Pickle the 'data' dictionary using the highest protocol available.
        pickle.dump(data, f, pickle.HIGHEST_PROTOCOL)
```

# 3. Image subsampling comparison

```python
# compare our 3 sampling methods for multiple images

naive_sampling_data = {}
average_sampling_data = {}
clean_sampling_data = {}
key_list = []
n=20

for k in random.sample(list(data.keys()), 5): # shuffle to get another set of image
  array = data.get(k)
  naive_sampling_data[k] = naive_sampling(array, n)
  average_sampling_data[k] = average_sampling(array, n)
  clean_sampling_data[k] = clean_sampling(array, n)

  key_list.append(k)
```

```python
fig, axes = plt.subplots(len(key_list), 4, figsize=(15,15))

for i in range(len(key_list)):
  k = key_list[i]
  axes[i][0].imshow(data[k]) # raw image
  axes[i][1].imshow(naive_sampling_data[k]) # naively sampled image
  axes[i][2].imshow(average_sampling_data[k]) # averagesampled image
  axes[i][3].imshow(clean_sampling_data[k]) # cleanly sampled image

# we see that the naive sampling has a high variance. Some small elements can constiture 1 element of the picture, and thus lead to a lot of dropped information.
# Average sampling helps more to distinguish part of the images, and averages the colors in each portion of the image.
# Average sampling and clean sampling methods are quite equivalent in term of results (but still clean sampling is better in term of details)
```

# 4. HOG feature comparison

```python
# compare our the hog features for different parameters
hog4 = {}
hog8 = {}
hog16 = {}
key_list = []

for k in random.sample(list(data.keys()), 3): # shuffle to get another set of image
  array = data.get(k)
  hog4[k] = compute_hog(array, 4)
  hog8[k] = compute_hog(array, 8)
  hog16[k] = compute_hog(array, 16)
  key_list.append(k)
```

```python
fig, axes = plt.subplots(len(key_list), 4, figsize=(20,20))

for i in range(len(key_list)):
  k = key_list[i]
  axes[i][0].imshow(data[k]) # raw image
  axes[i][1].imshow(hog16[k]) # 16 hog image
  axes[i][2].imshow(hog8[k]) # 8 hog image
  axes[i][3].imshow(hog4[k]) # 4 hog image

# with hog=4 we loose a lot of information (the geometric aspect of the image is a bit lost in non rectangular areas), but not so much from switching from 8 to 16. It would be interesting to see the impact on the prediction results.
```

```python
#Use raw data as input of CNN

# Train/Test split

X_train, X_test, Y_train, Y_test = train_test_split(X, Y, test_size=0.2)
del X # useful since X_train and X_test take additionnal space (which is valuable for the fitting of our model)
```

```python
from tensorflow.math import argmax, reduce_sum, equal, reduce_all
from tensorflow.linalg import matmul
from tensorflow import convert_to_tensor, reshape, tensordot, multiply, cast, float32
from tensorflow.nn import softmax_cross_entropy_with_logits

zero_tensor = cast(convert_to_tensor(0), float32)
def custom_loss(y_true, y_pred):
    #return K.mean(K.abs(y_true-y_pred)*class_weight)
    if not reduce_all(equal(y_true, y_pred)):
      return reduce_sum(multiply(weight_tensor, cast(reshape(y_true, [15]),float32)))
    else :
      return zero_tensor
```

```python
model = Sequential()
# convolutional layer
model.add(Conv2D(30, (5,5), input_shape=X_train.shape[1:]))
model.add(Activation("relu"))
model.add(Dropout(0.2))
#model.add(MaxPool2D(pool_size=(3,3)))

model.add(Conv2D(20, (4,4), input_shape=X_train.shape[1:]))
model.add(Activation("relu"))
model.add(Dropout(0.2))
#model.add(MaxPool2D(pool_size=(3,3)))

model.add(Conv2D(15, (3,3), input_shape=X_train.shape[1:]))
model.add(Activation("relu"))
model.add(Dropout(0.6))
model.add(MaxPool2D(pool_size=(5,5)))
model.add(Flatten())

# output layer : 15 labels to predict
model.add(Dense(15, activation='softmax'))


# compiling the sequential model
#loss1 = weightedLoss(CategoricalCrossentropy, class_weight)
#loss2 = custom_loss(CategoricalCrossentropy, class_weight)
loss3 = CategoricalCrossentropy
loss4 = custom_loss
loss5 = softmax_cross_entropy_with_logits
model.compile(loss=loss5, metrics=Accuracy(), optimizer="Adam")
#model.compile(loss=CategoricalCrossentropy, metrics = Accuracy(), optimizer="Adam")
#model.compile(loss=, metrics=Accuracy(), optimizer='adam')
```
